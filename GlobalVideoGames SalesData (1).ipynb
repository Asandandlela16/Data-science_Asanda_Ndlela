```python
print(df.isnull().sum()) #Checking for missing values 

df = df.dropna(subset=['Name', 'Global_Sales'])      #Drop the rows with null values

df['User_Score'] = df['User_Score'].fillna(0)     # fill the missing ratings with zero(0)

df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')    # convert user score to numeric
```

```python
df.to_csv("cleaned_video_game_sales.csv", index=False)
```
