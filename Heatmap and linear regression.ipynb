```python
#import the necessary libaries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Generate sample data
#creating a DatFrame with random data
np.random.seed(40)
data=pd.DataFrame({
    'feature1' :np.random.rand(100) ,
    'feature2' :np.random.rand(100) *2 ,
    'feature3' :np.random.rand(100) *3 ,
    'feature4' :np.random.rand(100) *4 ,
    'feature5' :np.random.rand(100) *5 ,
})

#calculation of the correlation matrix
corr_matrix=data.corr()

#ploting  the heatmap
plt.figure(figsize=(7,7))
heatmap=sns.heatmap(corr_matrix, annot=True , cmap="coolwarm" ,center=0)
plt.title("correlationof Heatmap")
plt.show
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score

np.random.seed(42)
x=2.5*np.random.rand(100,2)
y=5* x+np.random.rand(100,2)

data=pd.DataFrame({'feature':x.flatten() , 'Target' : y.flatten() })

x_train, x_test , y_train , y_test= train_test_split(x,y, test_size=0.3 , random_state=41)

model=LinearRegression()
model.fit(x_train , y_train)

y_pred=model.predict(x_test)

plt.figure(figsize=(8,6))
plt.scatter(x_test,y_test , color="purple" , label='Actual data')
plt.plot(x_test , y_test , color='red', linewidth=2, label="Regression line")
plt.xlabel('feature')
plt.ylabel('Target')
plt.title("Linear Regression Analysis")
plt.legend()
plt.show()

print('intercect(b0):' , model.intercept_[0])
print('slope(b1):', model.coef_[0][0])

mse=mean_squared_error(y_test , y_pred)
r2=r2_score(y_test , y_pred)
print("Mean Square Error", mse)
print("R-squared", r2)

```
